```python
import yfinance as yf
import pandas as pd

def yahoo_static(ticker: str) -> pd.DataFrame:
    t = yf.Ticker(ticker)
    info = t.info  # dict Yahoo

    data = {
        "Ticker": ticker,
        "Name": info.get("longName") or info.get("shortName"),
        "Description": info.get("longBusinessSummary"),
        "QuoteType": info.get("quoteType"),
        "Exchange": info.get("exchange"),
        "Currency": info.get("currency"),
        "Country": info.get("country"),
        "Sector": info.get("sector"),
        "Industry": info.get("industry"),
        "ISIN": info.get("isin"),
        "Website": info.get("website"),
    }

    return pd.DataFrame([data])
```

```python
df = yahoo_static("PUST.PA")
df
```

```python
import pandas as pd
import yfinance as yf

def yahoo_etf_static(ticker: str) -> pd.DataFrame:
    t = yf.Ticker(ticker)
    info = t.info

    inception_raw = info.get("fundInceptionDate")

    # Conversion Unix timestamp â†’ date
    if inception_raw:
        inception_date = pd.to_datetime(inception_raw, unit="s").date()
    else:
        inception_date = None

    data = {
        "Ticker": ticker,
        "FundName": info.get("longName"),
        "FundFamily": info.get("fundFamily"),
        "Category": info.get("category"),
        "TotalAssets": info.get("totalAssets"),
        "InceptionDate": inception_date,   # <- vraie date
        "Currency": info.get("currency"),
        "Exchange": info.get("exchange"),
    }

    return pd.DataFrame([data])

```

```python
df = yahoo_etf_static("PUST.PA")
df
```
